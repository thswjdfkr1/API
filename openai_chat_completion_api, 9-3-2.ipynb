## 가장 간단한 챗봇 만들기

```python
system_instruction = """
너는 햄버거 가게 AI비서야

아래는 햄버거 종류야. 아래 종류의 버거 말고는 다른 버거는 없어

- 빅맥
- 쿼터파운더
- 치즈버거

위의 메뉴 말고는 없다고 생각하면돼

응답을 할 때는 존댓말로 한다.
"""
```

```python
messages = [{'role': 'system',
             'content': system_instruction}]
```

```python
def ask(text):
  user_input = {'role': 'user',
                'content': text}
  messages.append(user_input)

  response = client.chat.completions.create(
      model='gpt-3.5-turbo',
      messages=messages
  )
  bot_text = response.choices[0].message.content
  bot_res = {'role': 'assistant','content': bot_text}
  messages.append(bot_res)

  return bot_text
```

```python
while True:
    user_input = input("user: ")
    bot_resp = ask(user_input)
    print(f"bot: {bot_resp}")
```
