## 예시 돌려보기

```python
import base64
import vertexai
from vertexai.preview.generative_models import GenerativeModel, Part, Image
from IPython.display import Video
```

```python
Video('/content/drive/MyDrive/fastcam/Part 9. AI API 활용/실습 자료/Chapter 4. Google AI API/data/sample.mp4', width=640)
```

```python
with open('/content/drive/MyDrive/fastcam/Part 9. AI API 활용/실습 자료/Chapter 4. Google AI API/data/sample.mp4', 'rb') as f:
  file_bin = f.read()
```

```python
video_encoded = base64.b64encode(file_bin)
```

```python
video = Part.from_data(data=video_encoded.decode('utf-8'), mime_type='video/mp4')
```

```python
model = GenerativeModel('gemini-2.0-flash')
```

```python
response = model.generate_content(
    [video,'이 영상을 간단하게 한문장으로 설명해줘'],
     generation_config={
         'max_output_tokens':2048,
         'temperature':0.4,
         'top_p':1,
         'top_k':32
    },
    stream=True,
)
```

```python
for res in response:
  print(res.text, end="")
```
